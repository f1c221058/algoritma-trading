```python
import yfinance as yf
import pandas as pd

tickers_list = ['A', 'AAL', 'AAPL', 'ABBV', 'ABT']

#Menenteukan Batas Data yang akan diambil  dan batas akhir datanya 
start_date = '2015-09-28'
end_date = '2030-09-30'

# Gunakan paket dari yfinance untuk mengunduh data agar lebih mudah 
# group_by='ticker'` kode ini akan menghasilkan struktur MultiIndex yang sama 
print(f"Mengunduh data untuk {tickers_list}...")
data = yf.download(tickers_list, start=start_date, end=end_date, group_by='ticker')

#Data yang dihasilkan akan memiliki struktur yang sama 
print("\n--- Data yang berhasil diunduh ---")
print(data)

print("\n--- Data spesifik untuk tanggal 2015-09-29 ---")
try:
    specific_date_data = data.loc['2015-09-29']
    print(specific_date_data)
except KeyError:
    print("Data untuk tanggal 2015-09-29 tidak ditemukan. Mungkin itu adalah hari libur bursa.")
```

```python
import pandas as pd
import numpy as np

# 1. Definisikan dates dan buat DataFrame awal (Contoh)
dates = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03'])

# Membuat DataFrame Contoh yang sesuai dengan nama kolom Anda
# Kita fokus pada AAL dan ABBV sesuai MultiIndex yang Anda coba buat
tickers = ['AAL', 'ABBV']
metrics = ['Open', 'High', 'Low', 'Close', 'Volume']
kolom = [f'{t}_{m}' for t in tickers for m in metrics]

# Membuat data acak untuk 3 baris (3 tanggal)
data_dummy = np.random.rand(len(dates), len(kolom)) * 100
# Mengganti kolom Volume menjadi integer (sesuai data volume saham)
for i in range(len(kolom)):
    if 'Volume' in kolom[i]:
        data_dummy[:, i] = np.round(data_dummy[:, i] * 1000).astype(int)

df_wide = pd.DataFrame(data_dummy, index=dates, columns=kolom)

print("DataFrame Awal (df_wide):\n", df_wide)
print("-" * 50)
```

```python
# Pisahkan nama kolom menjadi dua level MultiIndex: Ticker dan Metric
df_wide.columns = pd.MultiIndex.from_tuples(
    [tuple(col.split('_')) for col in df_wide.columns], 
    names=['Ticker', 'Metric']
)

# Gunakan .stack() untuk memindahkan level 'Metric' ke index baris
# Hasilnya adalah DataFrame MultiIndex (Date, Metric)
df_stacked = df_wide.stack(level='Metric')

# Opsional: Gunakan .stack() lagi untuk memindahkan 'Ticker' ke index baris
stacked_data = df_stacked.stack(level='Ticker').rename('Value') 
# .rename('Value') mengubah nama Series dari default menjadi 'Value'

print("\nSeries Hasil Akhir (stacked_data):\n", stacked_data)
```

```python

# --- 3. PERHITUNGAN VOLATILITAS GARMAN-KLASS ---
c = (2 * np.log(2) - 1)
gk_variance_daily = pd.DataFrame(index=df_unstacked.index)

for ticker in df_unstacked.columns.get_level_values('Ticker').unique():
    ohlc_data = df_unstacked[ticker]
    if all(col in ohlc_data.columns for col in ['Open', 'High', 'Low', 'Close']):
        term1 = (np.log(ohlc_data['High']) - np.log(ohlc_data['Low']))**2 / 2
        term2 = c * (np.log(ohlc_data['Close']) - np.log(ohlc_data['Open']))**2
        gk_variance_daily[ticker] = term1 - term2

print("\nVarians Harian Garman-Klass (5 baris pertama):")
print(gk_variance_daily.head())


# --- 4. PERHITUNGAN VOLATILITAS TAHUNAN ---
annualized_volatility = np.sqrt(gk_variance_daily.mean() * 252)

print("\nHasil Akhir: Volatilitas Tahunan Garman-Klass:")
print("="*40)
for ticker, vol in annualized_volatility.items():
    print(f"Saham {ticker}: {vol:.4f}  (atau {vol*100:.2f}%)")
print("="*40)
```
