```python
# 1. IMPOR LIBRARY YANG DIPERLUKAN
# Kita menggunakan zipline-reloaded dan pyfolio-reloaded
import pandas as pd
from datetime import datetime
import pyfolio as pf
from zipline import run_algorithm

# 1. Impor Fungsi Trading & Utility dari zipline.api
# Semua fungsi yang mengontrol perdagangan dan waktu harus diimpor dari 'api'
from zipline.api import (
    order_target_percent,
    symbol,
    get_datetime,
    # PENTING: set_commission dipindahkan ke api di banyak versi Zipline-reloaded
    set_commission
)

# 2. Impor Kelas Komisi dari zipline.finance.commission
# Hanya kelas (blueprint) untuk menentukan komisi yang diimpor dari sini
from zipline.finance.commission import PerTrade

# CATATAN PENTING:
# Impor eksplisit di atas hanya diperlukan jika Anda merujuk pada objek
# ini di luar definisi fungsi initialize/handle_data/before_trading_start.
# Saat algoritma dijalankan dengan 'run_algorithm', objek dari 'zipline.api'
# secara otomatis tersedia di dalam fungsi-fungsi tersebut.
```

```python
# 2. DEFINISI ALGORITMA

def initialize(context):
    """
    Fungsi ini dipanggil sekali di awal backtest.
    Digunakan untuk mengatur variabel dan parameter.
    """
    # Tentukan saham yang akan ditradingkan
    context.asset = symbol('AAPL')
    
    # Tentukan jendela rolling untuk moving average (90 hari)
    context.rolling_window = 90
    
    # Atur komisi perdagangan (misalnya, $5 per transaksi)
    set_commission(PerTrade(cost=5.0))

def handle_data(context, data):
    """
    Fungsi ini dipanggil untuk setiap bar data (misalnya, setiap hari).
    Berisi logika inti dari strategi trading.
    """
    # Ambil data historis harga penutupan untuk aset yang dipilih
    # dalam jangka waktu yang telah ditentukan.
    price_hist = data.history(context.asset, "close", context.rolling_window, "1d")
    
    # Logika trading:
    # Jika harga terakhir lebih tinggi dari rata-rata harga 90 hari, beli 100% modal.
    # Jika tidak, jual semua (posisi 0%).
    if price_hist[-1] > price_hist.mean():
        order_target_percent(context.asset, 1.0)
    else:
        order_target_percent(context.asset, 0.0)

def analyze(context, perf):
    """
    Fungsi ini dipanggil di akhir backtest.
    Digunakan untuk menganalisis hasil kinerja.
    """
    # Gunakan pyfolio-reloaded untuk membuat laporan kinerja
    # `perf` adalah dataframe yang berisi hasil backtest
    returns, positions, transactions = pf.utils.extract_rets_pos_txn_from_zipline(perf)
    
    print("\n--- Membuat Laporan Kinerja dengan Pyfolio-Reloaded ---")
    # Buat "tear sheet" (laporan lengkap) untuk analisis return
    pf.create_returns_tear_sheet(returns, benchmark_rets=None)

# 3. EKSEKUSI BACKTEST

```

```python
def main():
    # Tentukan periode backtest
    start_date = pd.to_datetime('2003-01-01', utc=True)
    end_date = pd.to_datetime('2018-01-01', utc=True)

    # Jalankan algoritma
    # Kita akan menggunakan bundle data 'csvdir-equities' yang harus kita daftarkan terlebih dahulu
    # (lihat penjelasan di bawah)
    results = run_algorithm(
        start=start_date,
        end=end_date,
        initialize=initialize,
        handle_data=handle_data,
        analyze=analyze,
        capital_base=10000,  # Modal awal
        data_frequency='daily',
        bundle='my-aapl-bundle' # Nama bundle data kita
    )
    
    # Cetak beberapa baris hasil terakhir
    print("\n--- Hasil Akhir Backtest ---")
    print(results.tail())

```

```python
if __name__ == '__main__':
    # Sebelum menjalankan, kita perlu mendaftarkan bundle data dari file CSV
    # Ini adalah langkah pengganti `bundle='quandl'`
    from zipline.utils.calendars import get_calendar
    from zipline.utils.run_algo import load_extensions

    # Fungsi untuk mendaftarkan bundle dari direktori CSV
    def register_csv_bundle():
        from zipline.data.bundles import csvdir_equities, register
        
        # Tentukan nama bundle, kalender, dan direktori tempat file CSV berada
        register(
            'my-aapl-bundle',
            csvdir_equities(
                ['daily'],
                './data',  # Direktori tempat Anda menyimpan file CSV
            ),
            calendar_name='NYSE', # Gunakan kalender NYSE
        )

    # Daftarkan bundle
    register_csv_bundle()
    
    # Jalankan fungsi utama
    main()
```
